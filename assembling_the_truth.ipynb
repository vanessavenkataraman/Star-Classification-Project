# Explanation

Again, these files had to be created in order to upload this to GitHub. This notebook is designed to assemble the truth set.

```python
import pandas as pd
```

```python
split_truth_1 = pd.read_csv("data_chunks/split_truth_1.csv")
split_truth_2 = pd.read_csv("data_chunks/split_truth_2.csv")
split_truth_3 = pd.read_csv("data_chunks/split_truth_3.csv")
split_truth_4 = pd.read_csv("data_chunks/split_truth_4.csv")
split_truth_5 = pd.read_csv("data_chunks/split_truth_5.csv")
```

```python
whole_truth_assembled = pd.concat([
    split_truth_1,
    split_truth_2,
    split_truth_3,
    split_truth_4,
    split_truth_5
]).sort_index()
```

```python
whole_truth_assembled
```

```python
whole_truth_assembled.to_csv("truth_set.csv", index = False)
```
